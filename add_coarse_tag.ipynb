### 기존 데이터셋에 대분류 태그 추가
기존 tag->는 fine_tag로 변경
- coarse_tag : 대분류 태그
- fine_tag : 세부 분류 태그

```python
import pandas as pd
```

```python
data = pd.read_json("./tagging_train_result.json")
```

```python
data
```

```python
data.tag.unique()
```

```python
purpose = ['문제 정의', '가설 설정', '기술 정의']
method = ['제안 방법', '데이터처리', '분석방법', '이론/모형']
result = ['성능/효과','후속연구']
```

```python
def get_coarse_tag(fine_tag):
    if fine_tag in purpose:
        return '연구 목적'
    if fine_tag in method:
        return '연구 방법'
    if fine_tag in result:
        return '연구 결과'
```

```python
for tag in data.tag:
    data['coarse_tag'] = get_coarse_tag(tag)
```

```python
data
```

```python
data.rename(columns = {'tag':'fine_tag'},inplace=True)
```
